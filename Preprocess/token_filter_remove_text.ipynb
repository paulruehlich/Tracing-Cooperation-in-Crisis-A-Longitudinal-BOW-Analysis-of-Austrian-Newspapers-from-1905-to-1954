```python
import pandas as pd
import numpy as np

df = pd.read_csv("final_ratios_ztransformed.csv")
```

```python
df.head()
```

```python
import ast

df['tokens'].apply(type).value_counts()
df['tokens'] = df['tokens'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

```python
df['token_count'] = df['tokens'].apply(lambda x: len(x) if isinstance(x, list) else 0)
df_subset = df[df['token_count'] >= 25]
```

```python
df_subset.head()
```

```python
df_subset.to_csv("filtered_25tokens_final_data.csv", index=False)
```

```python
columns_to_drop = ["text", "processed_text", "tokens", "freqs"]
missing = [col for col in columns_to_drop if col not in df_subset.columns]
print("Missing columns:", missing)
```

```python
df_subset_text = df_subset.drop(["text", "processed_text", "tokens", "freqs"], axis = 1)
```
